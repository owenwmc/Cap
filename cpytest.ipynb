```python
import numpy as np
import torch
from torch import nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

create a 2D convolutional layer, expecting an input with 16 channels
33 will produce 33 output channels (it has 33 different learned filters)
3 is the kernel size = create 3x3 filters to slide across input
stride = filter movement
stride = 2, filter moves 2 pixels at a time instead of 1
bias = false, layer won't learn bias terms
only conv filter weights will be trainable

torch.randn() generates values from std normal dist, shape follows pytorch convention for image-like data (batch size × channels × height × width)
this represents a batch of 20 images, each with 16 channels, where each channel is a 50×100 pixel grid
this matches the expected 16 input channels from the conv2d layer

when the convolution is applied batch size stays the same, channels become 33 as specified, spacial dimensions shrink because of stride
floor((input_size - kernel_size) / stride) + 1

this code creates a layer with random, untrained weights


```python
m = nn.Conv2d(16, 33, 3, stride=2, bias=False)

input = torch.randn(20, 16, 50, 100)
output = m(input)
print(f"Input shape: {input.shape}")
print(f"Output shape: {output.shape}")
```

```python
plt.imshow(input[0, 0].detach().numpy())
plt.show()
plt.imshow(output[0, 0].detach().numpy())
plt.show()
```
